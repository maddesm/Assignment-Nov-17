```python
# Part 1 : Bioinformatics Stronghold: 
## Assignment: Nov 17th
# FIBD (Fibonacci Rabbits dying)
with open('rosalind_fibd.txt') as input_data:
    n,m = map(int, input_data.read().split())
#initialization ( first rabbits)
Rabbits = [1]+[0]*(m-1)
# initial population in year 0
# baby rabbits are bunnies (new rabbits)
for year in range(1, n):
    Bunnies = 0
    # Get the number of Rabbits able to old enough to give birth.
    for j in range(1,m):
        Bunnies += Rabbits[(year-j-1)%m]
    # Bunnies replace the old rabbits who died.
    Rabbits[(year)%m] = Bunnies
Total_Rabbits = sum(Rabbits)
print(Total_Rabbits)
```

```python
#mRNA (Inferring mRNA from Protein)

from Bio.Data.CodonTable import standard_rna_table

# Load the standard codon table
codon_table = standard_rna_table.forward_table
stop_codons = standard_rna_table.stop_codons

# Create a dictionary to count codons per amino acid
rna_dict = {}

# Populate rna_dict with counts of codons for each amino acid
for codon, amino_acid in codon_table.items():
    if amino_acid not in rna_dict:
        rna_dict[amino_acid] = 0
    rna_dict[amino_acid] += 1

# Add stop codons to the dictionary
rna_dict['Stop'] = len(stop_codons)

with open('rosalind_mrna.txt') as input_data:
    protein = input_data.read().strip()

# Initialize with the number of stop codons
rna_num = rna_dict['Stop']

# Multiply by the number of codons for each amino acid in the protein
for p in protein:
    rna_num = (rna_num * rna_dict[p]) % 1000000

print(rna_num)
```

```python
# LIA (Independent Alleles/Mendell's Second Law)
from math import comb
#Computes the number of ways to choose i successful Aa Bb individuals from total_offspring
with open('rosalind_lia.txt') as input_data:
    k, N = map(int, input_data.read().split())

prob = 0
total_offspring = 2**k

for i in range(N, total_offspring + 1):
    prob += comb(total_offspring, i) * ((1/4.0)**i) * ((3/4.0)**(total_offspring - i))

print(prob)
```

```python
#Lcsm (Finding a Shared Motif)
# Read a FASTA file and return a list of sequences

def ReadFASTA(file_path):
    with open(file_path, 'r') as file:
        sequences = []
        current_seq = []
        for line in file:
            line = line.strip()
            if line.startswith('>'):
                if current_seq:
                    sequences.append(''.join(current_seq))
                current_seq = []
            else:
                current_seq.append(line)
        if current_seq:
            sequences.append(''.join(current_seq))
    return sequences

#Finds the longest common substring in a list of DNA sequences.
def LongestCommonSubstring(sequences):

    reference = sequences[0]  # Use the first sequence as the reference
    longest_substring = ''
    
    for i in range(len(reference)):
        for j in range(i + 1, len(reference) + 1):
            candidate = reference[i:j]
            if all(candidate in seq for seq in sequences):
                if len(candidate) > len(longest_substring):
                    longest_substring = candidate
    return longest_substring

if __name__ == '__main__':
    sequences = ReadFASTA('rosalind_lcsm.txt')
    result = LongestCommonSubstring(sequences)
    print(result)
```

```python
#PERM (Enumerating Gene Orders)

from itertools import permutations
from math import factorial

with open('rosalind_perm.txt') as input_file:
    n = int(input_file.read().strip())

# Generate all permutations of numbers 1 through n via permutation
perm_list = permutations(range(1, n + 1))

# Write to the output file each in in seperate line and also divided by space
with open('output_perm.txt', 'w') as output_file:
    # Write the total number of permutations (n!)
    output_file.write(f"{factorial(n)}\n")
    
print(factorial(n))
for perm in permutations(range(1, n + 1)):
    print(' '.join(map(str, perm)))
```

```python
#Cons(Consensus and Profile)

import numpy as np
from Bio import SeqIO 

# Parse the FASTA file
with open('rosalind_cons.txt') as input_file:
    dna_list = [str(record.seq) for record in SeqIO.parse(input_file, 'fasta')]

# Initialize the profile matrix
length = len(dna_list[0])
profile = np.zeros((4, length), dtype=int)
nucleotide_map = {'A': 0, 'C': 1, 'G': 2, 'T': 3}

# Fill the profile matrix
for dna in dna_list:
    for index, nucleotide in enumerate(dna):
        profile[nucleotide_map[nucleotide]][index] += 1

# Determine the consensus string
consensus = ''
nucleotide_order = ['A', 'C', 'G', 'T']
for i in range(length):
    most_common = np.argmax(profile[:, i])
    consensus += nucleotide_order[most_common]

output = [consensus]
for i, nucleotide in enumerate(nucleotide_order):
    output.append(f"{nucleotide}: {' '.join(map(str, profile[i]))}")

print('\n'.join(output)) 
```

```python
#Revp (Locating Restriction Sites)
# Function to find reverse palindromes in the DNA sequence
# Palindromes of length between 4 and 12

from Bio import SeqIO
from Bio.Seq import Seq

with open('rosalind_revp.txt') as input_file:
    dna = str(next(SeqIO.parse(input_file, 'fasta')).seq)

locations = []


for length in range(4, 13):  
    for index in range(len(dna) - length + 1):
        segment = dna[index:index + length]
        if segment == str(Seq(segment).reverse_complement()):
            locations.append(f"{index + 1} {length}")

print("\n".join(locations))
```

```python
#SPLC (RNA Splicing)

from Bio import SeqIO
from Bio.Seq import Seq

with open('rosalind_splc.txt') as input_file:
    sequences = [str(record.seq) for record in SeqIO.parse(input_file, "fasta")]
#initialiazation :
main_dna = sequences[0] #pre-mRNA is the first 
introns = sequences[1:] 
#get the introns out off the sequence
for intron in introns:
    main_dna = main_dna.replace(intron, '')

# Translate the mature DNA sequence to a protein sequence (via Translate function)
protein = Seq(main_dna).translate(to_stop=True)
print(protein)
```

```python
#TRAN (Transitions and Transversions)

from Bio import SeqIO

with open('rosalind_tran.txt') as input_file:
    sequences = [str(record.seq) for record in SeqIO.parse(input_file, "fasta")]

#Initialization :
dna1, dna2 = sequences[0], sequences[1]
purines = {'A', 'G'}
pyrimidines = {'C', 'T'}
transitions = 0
transversions = 0

# Comparing the two DNA sequences 
for base1, base2 in zip(dna1, dna2):
    if base1 != base2:
        if (base1 in purines and base2 in purines) or (base1 in pyrimidines and base2 in pyrimidines):
            transitions += 1
        else:
            transversions += 1

print(transitions / transversions)
```
